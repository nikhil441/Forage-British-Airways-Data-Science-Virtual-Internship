### Scraping data from Skytrax

If you visit [https://www.airlinequality.com] you can see that there is a lot of data there. For this task, we are only interested in reviews related to British Airways and the Airline itself.

If you navigate to this link: [https://www.airlinequality.com/airline-reviews/british-airways] you will see this data. Now, we can use `Python` and `BeautifulSoup` to collect all the links to the reviews and then to collect the text data on each of the individual review links.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 20
page_size = 100

reviews = []

# for i in range(1, pages + 1):
for i in range(1, pages + 1):

    print(f"Scraping page {i}")

    # Create URL to collect links from paginated data
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    # Collect HTML data from this page
    response = requests.get(url)

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        reviews.append(para.get_text())
    
    print(f"   ---> {len(reviews)} total reviews")
```

```python
df = pd.DataFrame()
df["reviews"] = reviews
df.head()
```

```python
df.to_csv("data/BA_reviews.csv")
```

Approach Based on Rules

The rule-based approach is a pragmatic method for text analysis that doesn't rely on training or utilizing machine learning models. This technique produces a set of predefined rules, often referred to as lexicons, which categorize the text as positive, negative, or neutral. Consequently, the rule-based approach is commonly known as the Lexicon-based approach.

Prominent lexicon-based approaches include TextBlob, VADER, and SentiWordNet.

Steps in Data Preprocessing:

  1.Text Cleaning
  2.Tokenization
  3.Enrichment through POS tagging
  4.Removal of stopwords
  5.Extraction of stem words

Step 1: Text Cleaning

```python
import re

# Define a function to clean the text
def clean_text(text):
    # Removes all special characters and numericals, leaving only the alphabets
    cleaned_text = re.sub('[^A-Za-z]+', ' ', str(text))
    return cleaned_text

# Cleaning the text in the 'reviews' column
# Applying the clean_text function to each entry in the 'reviews' column and creating a new column 'Cleaned Reviews'
df['Cleaned Reviews'] = df['reviews'].apply(clean_text)

# Displaying the first few rows of the DataFrame with the cleaned reviews
df.head()
```

Step 2: Tokenization
    
    Tokenization involves breaking down the text into smaller units known as tokens. This process can occur at either the sentence level (sentence tokenization) or the word level (word tokenization).

Step 3: Enrichment – POS Tagging

    Parts of Speech (POS) tagging transforms each token into a tuple in the form (word, tag). POS tagging is crucial for maintaining the context of words and plays a vital role in facilitating Lemmatization.

Step 4: Removal of Stopwords

    Stopwords are minimally informative words in English. It is essential to eliminate them as part of text preprocessing. The nltk library provides a list of stopwords for various languages.

Step 5: Obtaining Stem Words

    A stem represents a portion of a word responsible for its lexical meaning. Two popular techniques for obtaining root or stem words are Stemming and Lemmatization.

The primary distinction lies in the fact that Stemming often yields root words that lack clear meaning, as it simply truncates characters from the end. In contrast, Lemmatization produces meaningful root words, but it requires POS tags for the words.

    NLTK stands out as a prominent platform for developing Python programs focused on working with human language data. It offers user-friendly interfaces for over 50 corpora and lexical resources, including WordNet. Additionally, NLTK provides a suite of text processing libraries encompassing classification, tokenization, stemming, tagging, parsing, and semantic reasoning, along with wrappers for robust NLP libraries.

```python
import nltk

# The Punkt tokenizer segments a text into sentences using an unsupervised algorithm that constructs models for abbreviation words, collocations, and sentence-initial words.

# Downloading necessary resources
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk import pos_tag
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('wordnet')
from nltk.corpus import wordnet
```

```python
# The nltk.corpus package provides a range of corpus reader classes, allowing access to diverse corpora content.

# Downloading required resources
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

# Creating a part-of-speech (POS) tagger dictionary
pos_dict = {'J': wordnet.ADJ, 'V': wordnet.VERB, 'N': wordnet.NOUN, 'R': wordnet.ADV}

# Defining a function for tokenization, stopword removal, and POS tagging
def token_stop_pos(text):
    # Performing POS tagging on the tokenized words
    tags = pos_tag(word_tokenize(text))
    
    # Creating a new list to store tuples of words and their corresponding POS tags
    newlist = []
    
    # Iterating through each word and its tag, excluding stopwords
    for word, tag in tags:
        if word.lower() not in set(stopwords.words('english')):
            newlist.append(tuple([word, pos_dict.get(tag[0])]))
    
    return newlist

# Applying the tokenization, stopword removal, and POS tagging function to the 'Cleaned Reviews' column
df['POS tagged'] = df['Cleaned Reviews'].apply(token_stop_pos)

# Displaying the updated DataFrame with the POS-tagged information
df.head()
```

```python
# Performing Lemmatization to obtain stem words

from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

# Define a function for lemmatization
def lemmatize(pos_data):
    lemma_rew = " "
    
    # Iterating through words and their POS tags
    for word, pos in pos_data:
        if not pos:
            # If no POS tag is available, use the original word as the lemma
            lemma = word
            lemma_rew = lemma_rew + " " + lemma
        else:
            # Lemmatize using WordNetLemmatizer with the provided POS tag
            lemma = wordnet_lemmatizer.lemmatize(word, pos=pos)
            lemma_rew = lemma_rew + " " + lemma
    
    return lemma_rew

# Applying the lemmatization function to the 'POS tagged' column and creating a new column 'Lemma'
df['Lemma'] = df['POS tagged'].apply(lemmatize)

# Displaying the DataFrame with the 'Lemma' column containing the lemmatized results
df.head()
```

Conducting Sentiment Analysis with VADER

    VADER, an acronym for Valence Aware Dictionary and Sentiment Reasoner, not only discerns whether a statement is positive or negative but also provides information about the intensity of the emotional expression.

```python
# Importing the necessary library for sentiment analysis using VADER
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Function to calculate VADER sentiment
def vader_sentiment_analysis(review):
    # Obtaining the sentiment scores from VADER
    sentiment_scores = analyzer.polarity_scores(review)
    return sentiment_scores['compound']

# Applying the VADER sentiment analysis function to the 'Lemma' column and creating a new column 'Sentiment'
df['Sentiment'] = df['Lemma'].apply(vader_sentiment_analysis)

# Function to interpret sentiment categories based on the compound score
def interpret_sentiment(compound):
    if compound >= 0.5:
        return 'Positive'
    elif compound < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Applying the interpretation function to the 'Sentiment' column and creating a new column 'Analysis'
df['Analysis'] = df['Sentiment'].apply(interpret_sentiment)

# Displaying the DataFrame with the newly added 'Sentiment' and 'Analysis' columns
df.head()
```

```python
# Counting the occurrences of sentiment analysis categories
sentiment_counts = df['Analysis'].value_counts()
sentiment_counts
```

Word Cloud

    A Word Cloud, also known as Tag Cloud, is a visualization technique designed for displaying tags or keywords extracted from textual content. This method is commonly utilized to visually represent the prominence or frequency of words in a given context.

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)

def generate_wordcloud(data):
    # Create a WordCloud object with specified parameters
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=100,
        max_font_size=30,
        scale=3,
        random_state=1
    )

    # Generate the word cloud based on the provided data
    wordcloud = wordcloud.generate(str(data))

    # Display the generated word cloud using Matplotlib
    fig = plt.figure(1, figsize=(12, 12))
    
    # Turn off axis
    plt.axis('off')
    
    # Show the word cloud
    plt.imshow(wordcloud)
    plt.show()

# Generate and display the word cloud using the 'Lemma' column of the DataFrame
generate_wordcloud(df.Lemma)
```
